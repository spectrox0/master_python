Grafico petalo
Se hara primero con seaborn y luego con plotnine para comparar resultados

```python
from seaborn import FacetGrid

FacetGrid(df_iris, hue="species", height=5).map(plt.scatter, "petal_length", "petal_width").add_legend()
plt.show()
```

```python
scatter_plot = (ggplot(df_iris, aes(x="petal_length", y="petal_width", color="species"))
 + geom_point()
 + labs(title="Scatter plot of petal length vs petal width", x="Petal Length", y="Petal Width")
 + theme_bw()
)
print(scatter_plot)
```

Grafico Sepalo

```python
FacetGrid(df_iris, hue="species", height=5).map(plt.scatter, "sepal_length", "sepal_width").add_legend().set(title="Grafica para el sepalo con Seaborn")
```

```python
scatter_plot = (ggplot(df_iris, aes(x="sepal_length", y="sepal_width", color="species"))
 + geom_point()
 + labs(title="Grafica para el sepalo con Plotnine", x="Sepal Length", y="Sepal Width")
 + theme_bw()
)
print(scatter_plot)
```
