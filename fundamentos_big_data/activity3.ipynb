```python
# Pie chart


labels = ['A', 'B', 'C', 'D']
sizes = np.random.randint(50, 200, 4)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Ejemplo de Pie Chart')
plt.show()
```

```python
# Histogram


sample_data = np.random.randn(1000)
plt.hist(sample_data, bins=50, facecolor='blue', edgecolor='black', alpha=0.7)
plt.title('Ejemplo de Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True, alpha=0.6)
plt.show()
```

```python
df = pd.DataFrame({
    'X': np.random.rand(100),
    'Y': np.random.rand(100)
})

plt.scatter(df['X'], df['Y'], c='blue', marker='o', edgecolors='black')
plt.title('Ejemplo de Scatter Plot en 2D')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True, alpha=0.6)
plt.show()

# Scatter plot in 3D

df = pd.DataFrame({
    'X': np.random.rand(100),
    'Y': np.random.rand(100),
    'Z': np.random.rand(100)
})

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='r', marker='o')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Ejemplo de Scatter Plot en 3D')
plt.show()
```

## Fundamentos de Big Data 3
La estadística es una materia muy importante en Data Science. Existen algunos conceptos importantes, y a su vez, relacionados con Gráficos. Los “diagramas de cajas” son un elemento importante de los mismos.

```python
#Utilizaremos el dataset "Iris" para visualizar un diagrama de cajas de las longitudes de los pétalos de diferentes especies de iris:
import seaborn as sns
import matplotlib.pyplot as plt

# Cargando el dataset de Iris
iris = sns.load_dataset('iris')

# Mostrando el diagrama de cajas para la columna 'petal_length' agrupada por especies
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title('Longitudes de los pétalos de diferentes especies de Iris')
plt.show()
```

### Pregunta 2 – (hasta 1 punto)
¿ Qué son Q1, Q2, Q3 ?
Q1, Q2 y Q3 son cuartiles, que dividen un conjunto de datos en cuatro partes iguales:

Q1 (Primer cuartil): Es el valor medio entre el número más pequeño y la mediana (Q2) del conjunto de datos.
Q2 (Segundo cuartil o mediana): Es el valor medio del conjunto de datos.
Q3 (Tercer cuartil): Es el valor medio entre la mediana (Q2) y el número más grande del conjunto de datos.

### Pregunta 3 – (hasta 1 punto)
¿ Qué son los “outliers” ? ¿ son mostrados en estos diagramas de cajas?

Los "outliers" son valores atípicos o anomalías en un conjunto de datos. En el contexto de los diagramas de cajas, los outliers se representan típicamente como puntos fuera de los "bigotes" del gráfico. Sí, son mostrados en estos diagramas de cajas, y ayudan a identificar valores que se desvían significativamente del resto de la distribución.
